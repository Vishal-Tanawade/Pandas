# Concate

```python
india = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
india
```

```python
us = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
us
```

```python
df21=pd.concat([india,us])
df21
```

```python
df22=pd.concat([india,us],ignore_index=True)
df22
```

```python
df23=pd.concat([india,us],keys=["India1","US1"])
df23
```

```python
df23.loc["India1"]
```

```python
temperature_df = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
}, index=[0,1,2])
temperature_df
```

```python
windspeed_df = pd.DataFrame({
    "city": ["delhi","mumbai"],
    "windspeed": [7,12],
}, index=[1,0])
windspeed_df
```

```python
df24=pd.concat([temperature_df,windspeed_df])
df24
```

```python
df25=pd.concat([temperature_df,windspeed_df],axis=1)
df25
```

```python
s=pd.Series(["humid","Dry","Rain"],name="Event")
s
```

```python
df26=pd.concat([temperature_df,windspeed_df,s],axis=1)
df26
```

# Merge

```python
df31 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df31
```

```python
df32 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df32
```

```python
df33=pd.merge(df31,df32,on='city')
df33
```

```python
df34 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35, 38],
})
df34
```

```python
df35 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})
df35
```

```python
df36=pd.merge(df34,df35,on='city',how='outer')
df36
```

```python
df37=pd.merge(df34,df35,on='city',how='left')
df37
```

```python
df38=pd.merge(df34,df35,on='city',how='right')
df38
```

```python
df39=pd.merge(df34,df35,on='city',how='outer',indicator=True)
df39
```
